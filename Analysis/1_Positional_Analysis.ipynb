```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
PlayerDF = pd.read_csv('..\Player_Data\Cleaned_Player_Data.csv', index_col = 0)
```

```python
#Found range of advanced stats to set for x bounds in visualizations
limOBPM = (PlayerDF['OBPM'].min(),PlayerDF['OBPM'].max())
limDBPM = (PlayerDF['DBPM'].min(),PlayerDF['DBPM'].max())
limOWS = (PlayerDF['OWS'].min(),PlayerDF['OWS'].max())
limDWS = (PlayerDF['DWS'].min(),PlayerDF['DWS'].max())


PlayerDFGaurd = PlayerDF[(PlayerDF['Pos'] == 'PG') | (PlayerDF['Pos'] == 'SG')]
```

```python
PlayerDFGaurd
```

```python
'''
Gaurds are more likely to contribute to the offensive side of the ball. This is seen in the larger weight of salary in players with higher offensive win shares. For example,
a player with 3 deffensive win shares are moer likely to not be paid as much as a guard with 3+ offensive win shares. This shoes the leagues desire for offensive talented gaurds.
It is more common to see gaurds with high amounts of offensive win shares then deffensive win shares. 

The overall conclusion from these two graphs is that for gaurds, deffensive contributions are not as valued by teams. We can see that as the x values increase (Offensive metrics) there
is clear increase in salaries (darker points). This trend is much more extreme compared to a change in the y-axis where the trend of an increase in salary is not as clear. Additionally,
there is less vertical spread, showing the league trend that gaurds are primarily valued for offensive contributions rather than defense 
possibly due to lack of size and interior prescence. 

'''
fig, ax = plt.subplots(2,1, figsize = (20,20))

sns.scatterplot(data = PlayerDFGaurd, x = 'OWS', y = 'DWS', hue = 'YearSalaryM', palette='rocket_r', ax = ax[0])
ax[0].set_xlabel('Offensive Win Shares')
ax[0].set_ylabel('Defensive Win Shares')
ax[0].set_ylim(limDWS)
ax[0].set_xlim(limOWS)
ax[0].set_title('GUARDS')



sns.scatterplot(data = PlayerDFGaurd, x = 'OBPM', y = 'DBPM', hue = 'YearSalaryM', palette='rocket_r', ax = ax[1])
ax[1].set_xlabel('Offensive Box Plus Minus')
ax[1].set_ylabel('Deffensive Box Plus Minus')
ax[1].set_ylim(limDBPM)
ax[1].set_xlim(limOBPM)



plt.tight_layout()

```

```python
PlayerDFForward = PlayerDF[(PlayerDF['Pos'] == 'SF') | (PlayerDF['Pos'] == 'PF')]
PlayerDFForward
```

```python
'''
A similar trend for Forwards follows from the trends of Gaurds. Teams look for offensive stars and are willing to pay more for those who can bring wins and contribute heavily
to the offense side of the ball. Contrasting to the gaurds, there seems to be a larger spread of data, indicating that forwards are more versatile position, where some lean towards 
offense and some lean towards defense. 
'''
fig, ax = plt.subplots(2,1, figsize = (20,20))

sns.scatterplot(data = PlayerDFForward, x = 'OWS', y = 'DWS', hue = 'YearSalaryM', palette='rocket_r', ax = ax[0])
ax[0].set_title('FORWARDS')
ax[0].set_xlabel('Offensive Win Shares')
ax[0].set_ylabel('Defensive Win Shares')
ax[0].set_ylim(limDWS)
ax[0].set_xlim(limOWS)


sns.scatterplot(data = PlayerDFForward, x = 'OBPM', y = 'DBPM', hue = 'YearSalaryM', palette='rocket_r', ax = ax[1])
ax[1].set_xlabel('Offensive Box Plus Minus')
ax[1].set_ylabel('Deffensive Box Plus Minus')
ax[1].set_ylim(limDBPM)
ax[1].set_xlim(limOBPM)

plt.tight_layout()
```

```python
PlayerDFCenter = PlayerDF[PlayerDF['Pos'] == 'C']
PlayerDFCenter
```

```python
'''
The graphs of Centers seem to be unique to the other positional groupings. There is a much heavier emphasis on defense. Deffensive win shares seem to be paid more than offensive 
win shares for this position. Additionally, in the second graph, the points seem to be shifted up compared to the other positions. We can now see that as the 
y value increase (general defensive contributions) the salary increase. Additionally, many centers play power forward and many power forwards play center. This could account for 
some varying trends in the graphs of forwards and centers. This is because only one position is listed in the dataset, however players accumulate stats playing different positions. 
'''



fig, ax = plt.subplots(2,1, figsize = (20,20))

sns.scatterplot(data = PlayerDFCenter, x = 'OWS', y = 'DWS', hue = 'YearSalaryM', palette='rocket_r', ax = ax[0])
ax[0].set_title('CENTERS')
ax[0].set_xlabel('Offensive Win Shares')
ax[0].set_ylabel('Defensive Win Shares')
ax[0].set_ylim(limDWS)
ax[0].set_xlim(limOWS)


sns.scatterplot(data = PlayerDFCenter, x = 'OBPM', y = 'DBPM', hue = 'YearSalaryM', palette='rocket_r', ax = ax[1])
ax[1].set_xlabel('Offensive Box Plus Minus')
ax[1].set_ylabel('Deffensive Box Plus Minus')
ax[1].set_ylim(limDBPM)
ax[1].set_xlim(limOBPM)
```
