```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
PlayerDF = pd.read_csv('..\Player_Data\Cleaned_Player_Data.csv', index_col = 0)
```

```python
PlayerDFShooters = PlayerDF[(PlayerDF['3PAr'] > 0.5) & (PlayerDF['3P%'] > 0.4)]
PlayerDFScorers = PlayerDF[(PlayerDF['PTS'] > 20) & (PlayerDF['USG%'] > 25)]
PlayerDFFacilitators = PlayerDF[(PlayerDF['AST'] > 5) & (PlayerDF['PTS'] < 20)]
PlayerDFDefenders = PlayerDF[PlayerDF['STL'] + PlayerDF['BLK'] > 2]
```

```python
medSht = PlayerDFShooters['YearSalaryM'].median()
medScr = PlayerDFScorers['YearSalaryM'].median()
medFac = PlayerDFFacilitators['YearSalaryM'].median()
medDef = PlayerDFDefenders['YearSalaryM'].median()

fig, ax = plt.subplots(2,2, figsize = (10,10), sharex = True, sharey = True)
sns.histplot(data = PlayerDFShooters, x = 'YearSalaryM',ax = ax[0][0])
ax[0][0].axvline(medSht, color = 'red', linestyle = '--')
ax[0][0].set_title('High Volume Shooters')

sns.histplot(data = PlayerDFScorers, x = 'YearSalaryM',ax = ax[0][1])
ax[0][1].axvline(medScr, color = 'red', linestyle = '--')
ax[0][1].set_title('Scorers')

sns.histplot(data = PlayerDFFacilitators, x = 'YearSalaryM',ax = ax[1][0])
ax[1][0].axvline(medFac, color = 'red', linestyle = '--')
ax[1][0].set_title('Facilitators')

sns.histplot(data = PlayerDFDefenders, x = 'YearSalaryM',ax = ax[1][1])
ax[1][1].axvline(medDef, color = 'red', linestyle = '--')
ax[1][1].set_title('Defenders')

fig.suptitle("Salary Distributions Per Player Type")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize = (10,10), sharey = True, sharex = True)
sns.scatterplot(data = PlayerDFShooters, x = 'PER', y = 'YearSalaryM', ax = ax[0][0])
ax[0][0].set_title('High Volume Shooters')

sns.scatterplot(data = PlayerDFScorers, x = 'PER', y = 'YearSalaryM', ax = ax[0][1])
ax[0][1].set_title('Scorers')

sns.scatterplot(data = PlayerDFFacilitators, x = 'PER', y = 'YearSalaryM', ax = ax[1][0])
ax[1][0].set_title('Facilitators')

sns.scatterplot(data = PlayerDFDefenders, x = 'PER', y = 'YearSalaryM', ax = ax[1][1])
ax[1][1].set_title('Defenders')

for i in range(2):
    for j in range(2):
        ax[i][j].set_ylabel('')
        ax[i][j].set_xlabel('')

fig.suptitle('PER vs Salary(M): Player Classification')
plt.tight_layout()
plt.show()
```
